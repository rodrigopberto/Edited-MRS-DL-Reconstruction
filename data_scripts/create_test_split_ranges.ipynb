```python
def split_to_ranges(in_filename,out_filename):
    with h5py.File(in_filename) as hf:
        filenames = hf["filenames"][()]
        ppm = hf["ppm"][()]
        target_spectra = hf["target_spectra"][()]
        target_spectra_off = hf["target_spectra_off"][()]
        target_spectra_on = hf["target_spectra_on"][()]
        transient_specs = hf["transient_specs"][()]
    
    out_transient_specs = []
    out_filenames =[]
    out_ppms = []
    out_target_spectra = []
    out_target_spectra_on = []
    out_target_spectra_off = []
    out_ranges = []

    ranges = [(i*5,i*5+40) for i in range(25)]

    for i in range(ppm.shape[0]):
        for i_range in ranges:
            out_transient_specs += [transient_specs[i:i+1,:,:,i_range[0]:i_range[1]]]
            out_ranges += [f"{i_range[0]*2+1}-{i_range[1]*2}"]
            out_filenames+= [str(filenames[i]).split("'")[1]]
            out_ppms += [ppm[i:i+1]]
            out_target_spectra += [target_spectra[i:i+1]]
            out_target_spectra_off += [target_spectra_off[i:i+1]]
            out_target_spectra_on += [target_spectra_on[i:i+1]]
    
    print(out_ranges)



    with h5py.File(out_filename,"w") as hf:
        hf.create_dataset("filenames",data=out_filenames)
        hf.create_dataset("ppm",data=np.concatenate(out_ppms,axis=0))
        hf.create_dataset("target_spectra",data=np.concatenate(out_target_spectra,axis=0))
        hf.create_dataset("target_spectra_on",data=np.concatenate(out_target_spectra_on,axis=0))
        hf.create_dataset("target_spectra_off",data=np.concatenate(out_target_spectra_off,axis=0))
        hf.create_dataset("transient_specs",data=np.concatenate(out_transient_specs,axis=0))
        hf.create_dataset("ranges",data=out_ranges)
```

```python
split_to_ranges("splits/test_data_split_1.h5","split_ranges/test_data_range_split_1.h5")
```
