```python
import h5py
import numpy as np
```

```python
with h5py.File("data/split_ranges/test_data_range_split_2.h5") as hf:
    filenames = hf["filenames"][()]
    ppm = hf["ppm"][()]
    ranges = hf["ranges"][()]
    target_spectra = hf["target_spectra"][()]
    target_spectra_on = hf["target_spectra_on"][()]
    target_spectra_off = hf["target_spectra_off"][()]
    transient_specs = hf["transient_specs"][()]
    
```

```python
ranges
```

```python
new_filenames_list=[]
new_ppm_list=[]
new_ranges_list=[]
new_target_spectra_list=[]
new_target_spectra_on_list=[]
new_target_spectra_off_list=[]
new_transient_specs_list=[]

for i in range(filenames.shape[0]):
    if str(ranges[i]) in ["b'1-80'","b'81-160'","b'161-240'","b'241-320'"]:
        new_filenames_list+=[filenames[i:i+1]]
        new_ppm_list+=[ppm[i:i+1]]
        new_ranges_list+=[ranges[i:i+1]]
        new_target_spectra_list+=[target_spectra[i:i+1]]
        new_target_spectra_on_list+=[target_spectra_on[i:i+1]]
        new_target_spectra_off_list+=[target_spectra_off[i:i+1]]
        new_transient_specs_list+=[transient_specs[i:i+1]]

new_filenames = np.concatenate(new_filenames_list,axis=0)
new_ppm = np.concatenate(new_ppm_list,axis=0)
new_ranges = np.concatenate(new_ranges_list,axis=0)
new_target_spectra = np.concatenate(new_target_spectra_list,axis=0)
new_target_spectra_on = np.concatenate(new_target_spectra_on_list,axis=0)
new_target_spectra_off = np.concatenate(new_target_spectra_off_list,axis=0)
new_transient_specs = np.concatenate(new_transient_specs_list,axis=0)
```

```python
with h5py.File("data/split_ranges/test_data_range_split_2_v2.h5","w") as hf:
    hf.create_dataset("filenames",data=new_filenames)
    hf.create_dataset("ppm",data=new_ppm)
    hf.create_dataset("ranges",data=new_ranges)
    hf.create_dataset("target_spectra",data=new_target_spectra)
    hf.create_dataset("target_spectra_on",data=new_target_spectra_on)
    hf.create_dataset("target_spectra_off",data=new_target_spectra_off)
    hf.create_dataset("transient_specs",data=new_transient_specs)
```
