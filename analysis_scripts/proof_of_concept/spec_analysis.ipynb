```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
spec_folder = "../miccai/outputs"
filenames = os.listdir(spec_folder)
with h5py.File("../data/splits/test_data_split_1.h5") as hf:
    specs = {"target":hf["target_spectra"][()]}
    ppm = hf["ppm"][()]
for filename in filenames:
    with h5py.File(f"{spec_folder}/{filename}") as hf:
        specs[filename.split(".")[-2]] = hf["reconstruction"][()]
```

```python
ppm.shape
```

```python
idx = 20

fig,ax = plt.subplots(figsize=(10,8))
offset=0.005
for i,key in enumerate(specs):
    ax.plot(ppm[idx],specs[key][idx]-i*offset,label=key)


ax.set_xlim(1,5)
ax.invert_xaxis()
plt.legend()
plt.show()
```
