```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../eval/fit_ranges/size_increase.csv")
ref_df = pd.read_csv("../pre_results/full_ref_df.csv")

df = pd.merge(df,ref_df,on="filename",suffixes=("","_full"))

df["gaba_conc_cr_error"] = df["gaba_conc_cr"] - df["gaba_conc_cr_full"]
```

```python
filenames = list(df["filename"].unique())

fig,ax = plt.subplots()

n_figs = len(filenames)

for i in range(n_figs):
    filename=filenames[i]
    i_df = df[df["filename"]==filename].sort_values("size")
    
    ax.plot(i_df["size"],i_df["gaba_conc_cr"]/i_df["gaba_conc_cr_full"])

ax.set_ylim(0.5,1.5)
plt.show()
```

```python
filenames = list(df["filename"].unique())

fig,ax = plt.subplots()

n_figs = len(filenames)

for i in range(n_figs):
    filename=filenames[i]
    i_df = df[df["filename"]==filename].sort_values("size")
    
    ax.plot(i_df["size"],abs(i_df["gaba_conc_cr"]-i_df["gaba_conc_cr_full"]))

ax.set_ylim(0,0.1)
plt.show()
```

```python
sizes = list(df.sort_values("size")["size"].unique())

means=[]
medians=[]
stds=[]

fig,ax = plt.subplots()

n_figs = len(filenames)

for i in range(len(sizes)):
    size=sizes[i]
    i_df = df[df["size"]==size]
    mean = (i_df["gaba_conc_cr"]/i_df["gaba_conc_cr_full"]).mean()
    median = (i_df["gaba_conc_cr"]/i_df["gaba_conc_cr_full"]).median()
    std = (i_df["gaba_conc_cr"]/i_df["gaba_conc_cr_full"]).std()
    
    means.append(mean)
    medians.append(median)
    stds.append(std)

ax.plot(sizes,medians)
ax.plot(sizes,means)


ax.set_ylim(0.5,1.5)
plt.show()
```
